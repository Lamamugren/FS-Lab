```python
import pandas as pd 
import numpy as np
```

```python
df = pd.DataFrame({'Client': ['sara', 'saad', 'nora', 'mohammed'],
                   'Gender': ['Male', 'Female', 'Male', 'Female'],
                   'Marital Status': ['Married', 'Single', 'Widowed', 'Divorced'],
                   'Case Type': ['Criminal', 'Civil', 'Criminal', 'Civil'],
                   'Case Status': ['Open', 'Closed', 'Open', 'Closed']})


mapping = {'Male': 1, 'Female': 2, 'Married': 1, 'Single': 2, 'Widowed': 3, 'Divorced': 4,
           'Criminal': 1, 'Civil': 2, 'Open': 1, 'Closed': 2}

df

```

# Q1: Apply Ordinal encoding.

```python
df['Gender'] = df['Gender'].map(mapping)
df['Marital Status'] = df['Marital Status'].map(mapping)
df['Case Type'] = df['Case Type'].map(mapping)
df['Case Status'] = df['Case Status'].map(mapping)

df
```

# Q2: Apply One hot encoding.

```python
df = pd.get_dummies(df, columns=['Gender', 'Marital Status'])
df
```

# Q3: Apply Frequency Encoding. 

```python
frequency_encoding = df.groupby('Marital Status_1').size()
df['Marital Status_1'] = df['Marital Status_1'].map(frequency_encoding)
df
```

# Q4: Apply Binary Encoding. 

```python
binary_encoding = {'Married': 0, 'Single': 1, 'Widowed': 1, 'Divorced': 0}
df['Marital Status_2'] = df['Marital Status_2'].map(binary_encoding)
df
```
